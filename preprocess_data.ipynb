```python
colors = list(pd.Series(data['baseColour']).value_counts().index)
colors_num = list(pd.Series(data['baseColour']).value_counts())
plt.figure(figsize=(15, 6))
for i in range(len(colors)):
    plt.bar(colors[i], np.log2(colors_num[i]), color=np.random.randint(255, size=(1, 3)) / 255)
plt.xticks(colors, rotation=90)
plt.ylabel('Amount of Samples (log2)')
plt.title('Base Colour Distribution of Fashion Dataset')
plt.show()
plt.figure(figsize=(15, 6))
plt.show()
```
